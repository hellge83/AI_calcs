### Множество

Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.

```python
a = set('12345')
b = set('4567')
c = set()
```

```python
print(a.union(b))
print(b.union(a))
print(a.union(c))
print(c.union(a))
print(b.union(c))
print(c.union(b))
```

```python
print(a.intersection(b))
print(b.intersection(a))
print(a.intersection(c))
print(c.intersection(a))
print(b.intersection(c))
print(c.intersection(b))
```

```python
print(a.difference(b))
print(b.difference(a))
print(a.difference(c))
print(c.difference(a))
print(b.difference(c))
print(c.difference(b))
```

```python
print(a.symmetric_difference(b))
print(b.symmetric_difference(a))
print(a.symmetric_difference(c))
print(c.symmetric_difference(a))
print(b.symmetric_difference(c))
print(c.symmetric_difference(b))
```

### Последовательность

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

#### y = 2 ** x - x

```python
x = np.arange(1, 30)
y = (2 ** x) - x
print(y[4])
```

```python
plt.plot(x, y)
plt.show()
```

возрастает, неограничена

#### y = 1/(1-x)

```python
x = np.arange(2, 40)
y = 1/(1-x)
print(y[4])
```

```python
plt.plot(x, y)
plt.show()
```

возрастает, ограничена

#### y = (-1) ** x + np.sqrt(2 * x)

```python
x = np.arange(1, 40)
y = (-1) ** x + np.sqrt(2 * x)
print(y[4])
```

```python
plt.plot(x, y)
plt.show()
```

возрастает, неограничена

#### y = (-1) ** (2 * x) + 1/(x ** 2)

```python
x = np.arange(1, 40)
y = (-1) ** (2 * x) + 1/(x ** 2)
print(y[4])
```

```python
plt.plot(x, y)
plt.show()
```

убывает, ограничена

Найти 12-й член заданной неявно последовательности a[1] = 128, a[n+1] - a[n] = 6

```python
def series(n):
    if n == 1:
        return 128
    elif n > 1:
        return series(n - 1) + 6
    return series(n)
```

```python
print (series(12))
```

На языке Python предложить алгоритм вычисляющий численно предел с точностью 0.0000001 lim(n/((n!)** 1/n))

```python
from decimal import *
```

```python
def expr(n):
    return Decimal(n/np.power(np.math.factorial(n), 1/n))
```

```python
epsilon = 0.001
i = 2
while expr(i) - expr(i - 1) > epsilon:
    i+=1
print(expr(i))
```

```python
# на точности больше, чем 0.001 вызывается OverflowError: int too large to convert to float
```
